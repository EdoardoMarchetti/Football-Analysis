````python
match_csv = 'data\\Serie A\\2023\\all_matches_stats.csv'  
match_df = pd.read_csv(match_csv, index_col=0)
match_df.drop(columns=['Player', '#', 'Nation', 'Pos', 'Age','Team formation', 'Date', 'Matchweek', ], inplace=True)
````

````python
match_df.reset_index(inplace=True)
````

````python
index = ['Player ID','player_number','player_name','startin_XI','Team', 'Team ID','Match Id', 'Match Name']
vars = np.array(match_df.columns)[~match_df.columns.isin(index)]
````

````python
melt_df = match_df.melt(id_vars=index, value_vars=vars, var_name='stat')
````

````python
len(melt_df)
````

````python
for s in vars:
    melt_df.loc[melt_df.stat == s, 'perc'] = (melt_df.loc[melt_df.stat == s,'value'] /melt_df.loc[melt_df.stat == s,['Match Id','value']].groupby(by='Match Id').max().value[0])*100
````

````python
melt_df.fillna(value=0, inplace=True)
````

````python
filename = "data\\Serie A\\2023\\all_matches_stats_melted.csv"
````

````python
melt_df.to_csv(filename, index = False)
````
